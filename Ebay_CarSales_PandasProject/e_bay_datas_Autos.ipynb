#### Everything is OK
Let's check the brand column :

```python
brand_repartition = (autos["brand"].value_counts() / autos["brand"].value_counts().sum())*100
print(brand_repartition)
```

### We'll keep every brand over 3% of recurrency

```python
brand_top = brand_repartition[brand_repartition > 3].index
print(brand_top)
```

```python
avg_price_per_brand = {}
for brand in brand_top:
    brand_name = autos[autos["brand"] == brand]
    mean_price = brand_name["price"].mean()
    avg_price_per_brand[brand] = int(mean_price)
    
print(avg_price_per_brand)
print('\n')

a = sorted(avg_price_per_brand.items(), key=lambda x: x[1], reverse=True)    
print(a)

```

#### We see a gap in mean prices, between the top 3 expensive brands (Mercedes, Audi, BMW), and the other 4

### Let's convert this list in a panda series : 

```python
appb_series = pd.Series(avg_price_per_brand)
print(appb_series)

```

### And convert this series in a panda frame

```python
mean_price_df = pd.DataFrame(appb_series, columns=['mean_price'])
print(mean_price_df)
```

### Let's create a dataframe for mean mileage (odometer_km) :

```python

#odometer_km
#mean_price = {}
brand_mean_mileage = {}

for brand in brand_top:
    brand_name = autos[autos["brand"] == brand]
    mean_mileage = brand_name["odometer_km"].mean()
    brand_mean_mileage[brand] = int(mean_mileage)
    
print(brand_mean_mileage)
    

    
```
