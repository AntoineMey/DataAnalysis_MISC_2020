### **Exploring datas**

```python
from csv import reader

# Read the `artworks_clean.csv` file
opened_file = open('/content/drive/MyDrive/Dataquest/Data_Science_Questions_HackerNews/hacker_news_OK.csv')
read_file = reader(opened_file)
hn = list(read_file)
```

#### Deleting headers

```python
hn = hn[1:]
print(hn[0:5])
```

#### **Creating 3 lists for ASK, SHOW, and OTHERS**

```python
ask_posts = []
show_posts = []
other_posts = []

for row in hn:
    title = row[1].lower()
    
    if title.startswith('ask hn'):
        ask_posts.append(row)
    elif title.startswith('show hn'):
        show_posts.append(row)
    else : 
        other_posts.append(row)

print(len(ask_posts))
print(len(show_posts))
print(len(other_posts))


    
    
```

##### **Average Number of Comments for Ask and Show Posts**
 

```python
total_ask_comments = 0

for row in ask_posts: 
    ask_comments = int(row[4])
    total_ask_comments += ask_comments
    
print(total_ask_comments)
#print(len(ask_posts))

avg_ask_comments = total_ask_comments / len(ask_posts)
print('Avg number of comments under Ask posts: ' + str(avg_ask_comments)) 


print('\n')

total_show_comments = 0

for row in show_posts: 
    show_comments = int(row[4])
    total_show_comments += show_comments
    
print(total_show_comments)
avg_show_comments = total_show_comments / len(show_posts) 
print('Avg number of comments under Show posts: ' + str(avg_show_comments)) 


    
```

#### **Ask posts get more comments on average than Show posts, so we'll focus on them**
 


```python
import datetime as dt

result_list = []

for row in ask_posts:
    created_at = row[6]
    num_comments = int(row[4])
    
    result_list.append(
        [created_at, num_comments])

# print(len(result_list))
# print(result_list)


comments_by_hour = {}
counts_by_hour = {}
date_format = "%m/%d/%Y %H:%M"

    
for row in result_list:
    date = row[0]
    comment = row[1]
    time = dt.datetime.strptime(date, date_format).strftime("%H")
    if time in counts_by_hour:
        comments_by_hour[time] += comment
        counts_by_hour[time] += 1
    else:
        comments_by_hour[time] = comment
        counts_by_hour[time] = 1

comments_by_hour
```

#### **Checking the number of posts created in Ask, on average, by hour**  

```python
avg_by_hour = []

total_comments = 0
hours = comments_by_hour.values()
for item in hours :
    total_comments += item
print(total_comments)

print(counts_by_hour)

for hours in comments_by_hour:
    
    #val_post = counts_by_hour.values()
    #val_coms = comments_by_hour.values()
    
    avg_by_hour.append([hours, comments_by_hour[hours]/counts_by_hour[hours]])
    
avg_by_hour

    
    
```

#### **Sorting vaues of a list of list (avg_by_hour)** 

```python
swap_avg_by_hour = []
for item in avg_by_hour : 
    #avg_by_hour[item] = int(avg_by_hour[item])
    swap_avg_by_hour.append([item[1], item[0]])
    
x = sorted(swap_avg_by_hour, reverse = True)
topfive = x[0:5]
print('\n')
print("Top 5 Hours for Ask Posts Comments : ")
print('\n')

#FOR PYTHON 3+
#def utc_to_local(utc_dt):
    #return utc_dt.replace(tzinfo=timezone.utc).astimezone(tz=None)


for rows in topfive:
    datadate = rows[1]
    avg_comments = rows[0]
    datadate_obj = dt.datetime.strptime(datadate, "%H")
    
    #print(datadate_obj)
    #datadate_frm = datadate_obj.strftime("%H")   
    good_time = datadate_obj.strftime("%H:%M")
    output = """{} : {:.2f} average comments per post"""
    print(output.format(good_time, avg_comments))



```
